```python
import requests
```

```python
def pull_weather_data(lat, long, date_start, date_end, index):
    # Hard-code api-key
    meteostat_api_key = 'oP5skfxE'
    
    # Pull relevant station
    url_station = f'https://api.meteostat.net/v1/stations/nearby?lat={lat}&lon={long}&limit=5&key={meteostat_api_key}'
    r_station = requests.get(url_station)
    
    # Get weather data for relevant station
    for i in range(0,(len(r_station.json()['data']))):
        station_id = r_station.json()['data'][i]['id']
        print(station_id)
        url_history = f'https://api.meteostat.net/v1/history/daily?station={station_id}&start={date_start}&end={date_end}&key={meteostat_api_key}'
        r = requests.get(url_history)
        if r.json()['data'] == []:
            continue
        else:
            break
    # Create a list of the pulled back data
    weather = [index,station_id, 
               r.json()['data'][0]['temperature_max'], 
               r.json()['data'][0]['windspeed'], 
               r.json()['data'][0]['peakgust']]
    
    # Create a dataframe object to send out the data
    df = pd.DataFrame([weather])
    df.columns = ['index','station_id','temperature_max','windspeed','peakgust']

    return df
    
```
